CHIA DỮ LIỆU 70% TRAIN, 20% VALIDATION, 10% TEST

```python
features = [
    # 'ema_9', 
    # 'ema_25',
    # 'ema_50',
    # 'sma_9', 
    # 'sma_25',
    # 'sma_50',
    'macd',
    # 'macd_signal',
    'macd_diff',
    'sma_macd_diff_5',
    'sma_macd_diff_10', 
    'sma_macd_diff_sub',
    'rsi_14',
]

X = df[features]
y = df[Y_TYPE]

# Chia dữ liệu: 70% train, 30% test
# Bước 1: Chia 10% test
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, shuffle=False
)

# Bước 2: Chia 90% còn lại thành 77.8% train và 22.2% validation 
# (tương đương 70% và 20% của toàn bộ dữ liệu)
# X_train, X_val, y_train, y_val = train_test_split(
#     X_temp, y_temp, test_size=0.2222, random_state=42, shuffle=False
# )
```

GRADIENT BOOST PARAMS

```python
# gb_model = GradientBoostingClassifier(
#     n_estimators=200,         
#     learning_rate=0.01,       
#     max_depth=10,              
#     subsample=0.4,            # tỉ lệ mẫu dùng để huấn luyện mỗi cây (giúp tránh overfitting)
#     min_samples_split=5,      # số mẫu tối thiểu để chia node
#     min_samples_leaf=3,       # số mẫu tối thiểu ở lá
#     max_features='sqrt',      # số đặc trưng tối đa mỗi cây được dùng 
#     random_state=42           
# )

# gb_model.fit(X_train_low60_smote, y_train_low60_smote)
```

```python
gb_model = HistGradientBoostingClassifier(
    max_iter=1000,
    learning_rate=0.01,
    min_samples_leaf=5,
    max_leaf_nodes=32,
    l2_regularization=0.1,
    random_state=42,
    verbose=0,
    early_stopping=True,    
    n_iter_no_change=20,       
    validation_fraction=0.1        # dùng một phần nhỏ dữ liệu để xác thực
)

# Tính sample_weight để xử lý mất cân bằng nhãn
sample_weight = compute_sample_weight(class_weight='balanced', y=y_train)

gb_model.fit(X_train, y_train, sample_weight=sample_weight)

# Giả sử model của bạn là `model` và muốn lưu với tên theo `Y_TYPE`
# model_path = f'F:/dataML/models/GradientBoosting/{dataset_name}/{todaystr}/maxhigh/GB_{Y_TYPE}.joblib'
# os.makedirs(f"F:/dataML/models/GradientBoosting/{dataset_name}/{today_str}/maxhigh", exist_ok=True)
# joblib.dump(rf_model, model_path)
```

```python
print("Train accuracy:", gb_model.score(X_train, y_train))
print("Test accuracy:", gb_model.score(X_test, y_test))
```

ĐÁNH GIÁ ĐỘ CHÍNH XÁC

```python
# Dự đoán và tính độ chính xác
y_pred_gb = gb_model.predict(X_test)
accuracy_gb = accuracy_score(y_test, y_pred_gb)
print(f"Độ chính xác của mô hình Gradient Boosting: {accuracy_gb * 100:.2f}%")

# Lấy báo cáo phân loại dưới dạng dict
report = classification_report(y_test, y_pred_gb, output_dict=True)

# In báo cáo dạng phần trăm 
print("\nBáo cáo phân loại (%):")
print(f"{'Label':<12} {'Precision':>12} {'Recall':>12} {'F1-score':>12} {'Support':>10}")

def format_percent(value):
    return f"{value * 100:.2f}%"

for label, metrics in report.items():
    if isinstance(metrics, dict):
        print(f"{label:<12} {format_percent(metrics['precision']):>12} {format_percent(metrics['recall']):>12} {format_percent(metrics['f1-score']):>12} {int(metrics['support']):>10}")
    elif label == "accuracy":
        print(f"{label:<12} {'-':>12} {'-':>12} {format_percent(metrics):>12} {int(report['macro avg']['support']):>10}")
```

KẾT QUẢ

```python
# Tính TP, FP cho nhãn 1
tp_1 = np.sum((y_test == 1) & (y_pred_gb == 1))  # True Positive
fp_1 = np.sum((y_test == 0) & (y_pred_gb == 1))  # False Positive
total_predicted_1 = tp_1 + fp_1  # Tổng số lần mô hình dự đoán là nhãn 1

# Precision cho nhãn 1
precision_1 = tp_1 / total_predicted_1 if total_predicted_1 > 0 else 0

# Dữ liệu biểu đồ
labels = ['Tổng dự đoán nhãn 1', 'Dự đoán đúng (TP)', 'Dự đoán sai (FP)']
values = [total_predicted_1, tp_1, fp_1]
colors = ['blue', 'green', 'red']

# Vẽ biểu đồ
plt.figure(figsize=(8, 6))
bars = plt.bar(labels, values, color=colors)

# Thêm số liệu trên đầu cột
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 50, f'{int(yval)}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# Tiêu đề biểu đồ
plt.title(f'So sánh: Precision nhãn 1 = {precision_1:.2f}', fontsize=13)
plt.ylabel('Số lượng')
plt.ylim(0, max(values) * 1.2)
plt.tight_layout()
plt.show()
```

```python
# Ma trận nhầm lẫn
cm_gb = confusion_matrix(y_test, y_pred_gb)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_gb, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['Nhãn 0', 'Nhãn 1'],
            yticklabels=['Nhãn 0', 'Nhãn 1'])
plt.xlabel('Dự đoán')
plt.ylabel('Thực tế')
plt.title('Gradient Boosting')
plt.show()
```

```python
# Tính permutation importance cho HistGradientBoosting model (gb_model)
result = permutation_importance(
    gb_model,
    X_train,
    y_train,
    n_repeats=8,
    random_state=42,
    n_jobs=12
)

# Chuẩn hóa giá trị để tổng = 1
gb_importance_raw = pd.Series(result.importances_mean, index=features)
gb_importance = gb_importance_raw / gb_importance_raw.sum()
gb_importance = gb_importance.sort_values()

# Vẽ biểu đồ 
plt.figure(figsize=(8, 6))
gb_importance.plot(kind='barh', color='skyblue')
plt.title('Tầm quan trọng của đặc trưng - HistGradientBoosting (Hoán vị)')
plt.xlabel('Tầm quan trọng đã chuẩn hóa')
plt.ylabel('Đặc trưng')
plt.tight_layout()
plt.show()
```
